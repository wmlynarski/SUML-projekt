```python
import sqlite3
import pandas as pd

df = pd.read_csv('../datasets/Housing.csv')

display(df.head())
```

```python
print(df.info())

display(df.describe().round(2))
```

```python
df['price'] = (df['price'] * 0.039).round(2)

df['area'] = (df['area'] * 0.092903).round(2)
bool_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

for col in bool_columns:
    # Używamy .map, aby zamienić 'yes'/'no' na 1/0
    df[col] = df[col].map({'yes': 1, 'no': 0})

df['area'] = (df['area'] * 0.092903).round(2)

# 3. Mapowanie statusu umeblowania
furnishing_map = {'unfurnished': 0, 'semi-furnished': 1, 'furnished': 2}
df['furnishingstatus'] = df['furnishingstatus'].map(furnishing_map)
```

```python
db_path = '../datasets/housing.db'
conn = sqlite3.connect(db_path)

df.to_sql('apartments', conn, if_exists='replace', index=False)

conn = sqlite3.connect('../datasets/housing.db')
df.to_sql('apartments', conn, if_exists='replace', index=False)
conn.close()
```
